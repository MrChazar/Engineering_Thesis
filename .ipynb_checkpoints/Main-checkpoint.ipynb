## Funkcja Celu
Dla tego przykładu funkcja celu:
$$
\min \left( \sum_{i \in L} \sum_{n \in M} r_{in} \cdot x_{in} + \sum_{i \in L} \sum_{n \in M} c_i \cdot x_{in} \right)
$$

Będzie wyglądać:
$$
45.31*x_{11} + 7000*x_{12} + 488*x_{21} + 7300*x_{22} + 1000*x_{31} \\ + 6100*x_{32} + 2*x_{11} + 2*x_{12} + 4*x_{21} + 4*x_{22} + 0*x_{31} + 0*x_{32}
$$


Gdybyśmy zapomnieli o naszych ograniczeniach to macierz Q by wyglądała teraz tak wiedząc że
zmienne liniowe możemy zapisać jako kwadratowe

x_j = x_j^2
$$

wzór qubo:
$$
\min y = x^T Q x

## Ograniczenia
W problemie występują dane ograniczenia

$$
\sum_{i \in L} x_{i,n} = 1 \quad \forall n \in M
$$

$$
\alpha \cdot \sum_{n \in M} x_{i,n} \cdot f_{n} \leq v_i \quad \forall i \in L
$$

$$
\sum_{i \in L} c_i \cdot x_i \leq p
$$

$$
x_{i,n} = 0 \quad \text{jeśli } r_{i,n} > d
$$


Aby przekształcić ogranicznia możemy skorzystać z przekształceń w kary które wyglądają tak:
| Klasyczne Ograniczenie          | Odpowiadająca kara                          |
|-------------------------------|---------------------------------------------|
|  x + y <= 1             | P(xy)                                  |
|  x + y >= 1             |  P(1 - x - y + xy)                      |
|  x + y = 1               |  P(1 - x - y + 2xy)                     |
| x <= y               |  P(x - xy)                            |
|  x_1 + x_2 + x_3 <= 1  | P(x_1x_2 + x_1x_3 + x_2x_3)           |
|  x = y                   | P(x + y - 2xy)                         |


A więc dla ograniczenia:
$$ \sum_{i \in L} x_{i,n} = 1 \quad \forall n \in M $$
- Każdy obiekt $n$ musi być przypisany do dokładnie jednego schronu $i$

Które możemy rozpisać jako:
$$ x_{11} + x_{21} + x_{31} = 1 $$
$$ x_{12} + x_{22} + x_{32} = 1 $$

Zapisujemy jako karę: (czemu akurat taka kara ? sprawdź strona 28 dokument QUBO)
$$ P*(x_{11} + x_{21} + x_{31} - 1)^2 + P*(x_{12} + x_{22} + x_{32} - 1)^2 $$




$$
\alpha \cdot \sum_{n \in M} x_{in} \cdot f_{n} \leq v_i \quad \forall i \in L
$$
- Schron $i$ może obsłużyć tylko tylu mieszkańców $f_{n}$ obiektów $n$ biorąc pod uwagę stopień pokrycia $\alpha$, ile wynosi jego pojemność $v_{i}$

Które możemy rozpisać jako:
$$
0{,}1 \cdot (x_{11} \cdot 100 + x_{12} \cdot 155) \leq 100
$$

$$
0{,}1 \cdot (x_{21} \cdot 100 + x_{22} \cdot 155) \leq 50
$$

$$
0{,}1 \cdot (x_{31} \cdot 100 + x_{32} \cdot 155) \leq 94
$$
Zapisujemy jako karę:
$$ P*(100*x_{11} + 155x_{12} - 1000)^2 + P*(100*x_{11} + 155x_{12} - 500)^2 + P*(100*x_{11} + 155x_{12} - 940)^2  $$



$$
 \sum_{i \in L} \sum_{n \in M} c_i \cdot x_{in} \leq p
$$
- Koszt budowy $c_{i}$ nie może przekroczyć budżetu $p$

Które możemy rozpisać jako:
$$
2x_{11} + 2x_{12} + 4x_{21} + 4x_{22} + 0x_{31} + 0x_{32} \leq 4
$$

Zapisujemy jako karę:
$$
P*(2x_{11} + 2x_{12} + 4x_{21} + 4x_{22} + 0x_{31} + 0x_{32} - 4)^2
$$



$$
x_{in} = 0 \quad \text{jeśli } r_{i,n} > d
$$
-  Przypisania $x_{i,n}$ tylko tam, gdzie odległość nie przekracza $d$

Które możemy rozpisać jako:
$$ x_{12}=x_{22}=x_{32}=0 $$

Zapisujemy jako karę:
$$
P*(x_{12}+x_{22}+x_{32})^2
$$

# Zapisanie Qubo Python


## przykładowe dane

```python
s = 2 # dostępne lokalizacje nowe
e = 1 # istniejace juz schrony
h = 2 # liczba obiektów mieszkalnych
p = 4 # budżet na schrony
alpha = 0.1 # współczynnik pokrycia mieszkań
d = 3000 # Maksymalna dozwolona odległość
P = 10  # kara (dostosuj eksperymentalnie)

f = [100, 155] # Liczba mieszkańców w obiekcie mieszkalnym n
v = [100, 50, 94] # Pojemność schronu i
c = [2, 4, 0] # Koszty wybudowania schronu i

# Lokalizacje schrony i
L = [[51.103504, 17.086370],
    [51.100217, 17.082551],
    [51.100390, 17.099490]]

# Lokalizacje obiekty mieszkalne n
M = [[51.103910, 17.086310],
    [51.105504, 17.186370]]

# Odległości r_{i,n} w kilometrach
r = [[geodesic(l, m).meters for m in M] for l in L]
r = np.array(r)

# Zmienna decyzyjna jeśli schron
x = [[1 for _ in range(h)] for _ in range(s + e)]

# Inicjalizacja QUBO
N = (s + e) * h
Q = np.zeros((N, N))

def idx(i, n):
    return i * h + n
```

Funkcja celu: $$ \min \left( \sum_{i \in L} \sum_{n \in M} r_{in} \cdot x_{in} + \sum_{i \in L} \sum_{n \in M} c_i \cdot x_{in} \right) $$

```python
for i in range(s + e):
    for n in range(h):
        cost = c[i] + r[i][n]
        Q[idx(i, n), idx(i, n)] += cost
Q
```

Ograniczenia:
$$
\sum_{i \in L} x_{i,n} = 1 \quad \forall n \in M
$$

```python
for n in range(h):
    for i in range(s + e):
        for j in range(s + e):
            Q[idx(i, n), idx(j, n)] += P if i == j else 2 * P
    Q[idx(i, n), idx(i, n)] -= P
Q
```

$$
\alpha \cdot \sum_{n \in M} x_{i,n} \cdot f_{n} \leq v_i \quad \forall i \in L
$$

```python
for i in range(s + e):
    for n1 in range(h):
        for n2 in range(h):
            Q[idx(i, n1), idx(i, n2)] += P * alpha**2 * f[n1] * f[n2]
    for n in range(h):
        Q[idx(i, n), idx(i, n)] -= 2 * P * alpha * f[n] * v[i]
    Q[idx(i, i % h), idx(i, i % h)] += P * v[i]**2
Q
```

$$
\sum_{i \in L} c_i \cdot x_i \leq p
$$

```python
for i in range(s + e):
    for n1 in range(h):
        for n2 in range(h):
            Q[idx(i, n1), idx(i, n2)] += P * c[i] * c[i]
    for n in range(h):
        Q[idx(i, n), idx(i, n)] -= 2 * P * c[i] * p
Q[0, 0] += P * p**2  # dodanie P*p^2
Q
```

$$
x_{i,n} = 0 \quad \text{jeśli } r_{i,n} > d
$$


```python
for i in range(s + e):
    for n in range(h):
        if r[i][n] > d:
            Q[idx(i, n), idx(i, n)] += P
Q
```
