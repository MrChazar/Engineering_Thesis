## Ograniczenia
W problemie występują dane ograniczenia

$$
\sum_{i \in L}  x_{i,n} = 1 \quad \forall n \in M
$$

$$
\alpha \cdot \sum_{n \in M} x_{i,n} \cdot f_{n} \leq v_i \quad \forall i \in L
$$

$$
\sum_{i \in L} c_i \cdot x_i \leq p
$$

$$
x_{i,n} = 0 \quad \text{jeśli } r_{i,n} > d
$$


Aby przekształcić ogranicznia możemy skorzystać z przekształceń w kary które wyglądają tak:
| Klasyczne Ograniczenie          | Odpowiadająca kara                          |
|-------------------------------|---------------------------------------------|
|  x + y <= 1             | P(xy)                                  |
|  x + y >= 1             |  P(1 - x - y + xy)                      |
|  x + y = 1               |  P(1 - x - y + 2xy)                     |
| x <= y               |  P(x - xy)                            |
|  x_1 + x_2 + x_3 <= 1  | P(x_1x_2 + x_1x_3 + x_2x_3)           |
|  x = y                   | P(x + y - 2xy)                         |

Zaś jeśli są nierówności to możemy skorzystać z tzw. *slack variables*:

Przykład z nimi:

$$
3x_0 + 11x_1 + 14x_2 + 19x_3 + 5x_4 \le 26
$$

$$
3x_0 + 11x_1 + 14x_2 + 19x_3 + 5x_4 + S_1 = 26 \tag{4}
$$

gdzie:
$$
S = \sum_{k=0}^{N-1} 2^k s_k,
$$

$$
N = \lfloor\log_2(\max S)\rfloor + 1.
$$


czyli:
$$
S_1 = 2^0 s_{01} + 2^1 s_{11} + 2^2 s_{21} + 2^3 s_{31} + 2^4 s_{41}
$$

Funkcja karna:
$$
p(x,s) = \lambda \left(3x_0 + 11x_1 + 14x_2 + 19 x_3 + 5x_4 + x_5 + 2 x_6 + 4x_7 + 8 x_8 + 16 x_9 - 26\right)^2 \tag{5}
$$


### A więc dla ograniczenia #1:
$$ \sum_{i \in L} x_{i,n} = 1 \quad \forall n \in M $$
- Każdy obiekt $n$ musi być przypisany do dokładnie jednego schronu $i$

Które możemy rozpisać jako:
$$ x_{11} + x_{21} + x_{31} = 1 $$
$$ x_{12} + x_{22} + x_{32} = 1 $$

Zapisujemy jako karę: (czemu akurat taka kara ? sprawdź strona 28 dokument QUBO)
$$ P_1*(x_{11} + x_{21} + x_{31} - 1)^2 + P_1*(x_{12} + x_{22} + x_{32} - 1)^2 $$




### Ograniczenie #2:
$$
\alpha \cdot \sum_{n \in M} x_{in} \cdot f_{n} \leq v_i \quad \forall i \in L
$$
- Schron $i$ może obsłużyć tylko tylu mieszkańców $f_{n}$ obiektów $n$ biorąc pod uwagę stopień pokrycia $\alpha$, ile wynosi jego pojemność $v_{i}$

Które możemy rozpisać jako:
$$
0{,}1 \cdot (x_{11} \cdot 100 + x_{12} \cdot 155) \leq 100
$$

$$
0{,}1 \cdot (x_{21} \cdot 100 + x_{22} \cdot 155) \leq 50
$$

$$
0{,}1 \cdot (x_{31} \cdot 100 + x_{32} \cdot 155) \leq 94
$$
Stosujemy slack variable i zapisujemy:
$$
    (x_{11} \cdot 10 + x_{12} \cdot 15.5 + S_{1}) = 100
$$
$$
    (x_{21} \cdot 10 + x_{22} \cdot 15.5 + S_{2}) = 50
$$
$$
    (x_{31} \cdot 10 + x_{32} \cdot 15.5 + S_{3}) = 94
$$
gdzie dla każdego S:
$$
S = \sum_{k=0}^{N-1} 2^k s_k,
$$
$$Dla S_{1} N=7  S_{2} N=6  S_{3} N=7 $$
$$ S_{1}:
P_2*(x_{11} \cdot 10 + x_{12} \cdot 15.5  + s_{10} + s_{11}*2 + s_{12}*4 + s_{13}*8 + s_{14}*16 + s_{15}*32 + s_{16}*64 - 100)
$$
$$ S_{2}:
P_2*(x_{21} \cdot 10 + x_{22} \cdot 15.5  + s_{20} + s_{21}*2 + s_{22}*4 + s_{23}*8 + s_{24}*16 + s_{25}*32  - 50)
$$
$$ S_{3}:
P_2*(x_{31} \cdot 10 + x_{32} \cdot 15.5  + s_{30} + s_{31}*2 + s_{32}*4 + s_{33}*8 + s_{34}*16 + s_{35}*32 + s_{36}*64 - 94)
$$






### Ograniczenie #3:
$$
 \sum_{i \in L} \sum_{n \in M} c_i \cdot x_{in} \leq p
$$
- Koszt budowy $c_{i}$ nie może przekroczyć budżetu $p$

Które możemy rozpisać jako:
$$
P_3*(2x_{11} + 2x_{12} + 4x_{21} + 4x_{22} + 0x_{31} + 0x_{32} + S_{4} - 4)

$$

$$

Które możemy rozpisać jako:
$$ x_{12}=x_{22}=x_{32}=0 $$

# Zapisanie Qubo Python


## przykładowe dane

```python
import pandas as pd
from geopy.distance import geodesic # do obliczenia odległości
import numpy as np
import math

s = 2 # dostępne lokalizacje nowe
e = 1 # istniejace juz schrony
h = 2 # liczba obiektów mieszkalnych

alpha = 0.1 # współczynnik pokrycia mieszkań
d = 3000 # Maksymalna dozwolona odległość

P = [2, 2, 2]  # kara (dostosuj eksperymentalnie)
f = [100, 155] # Liczba mieszkańców w obiekcie mieszkalnym n

c = [2, 4, 0] # Koszty wybudowania schronu i

p = 4 # budżet na schrony
v = [100, 50, 94] # Pojemność schronu i

slack_sizes = [math.floor(math.log2(vi)) + 1 for vi in v] # za pojemności
slack_sizes.append(math.floor(math.log2(p)) + 1 ) # za budżet

# Lokalizacje schrony i
L = [[51.103504, 17.086370],
    [51.100217, 17.082551],
    [51.100390, 17.099490]]

# Lokalizacje obiekty mieszkalne n
M = [[51.103910, 17.086310],
    [51.105504, 17.186370]]

# Odległości r_{i,n} w kilometrach
r = [[geodesic(l, m).kilometers for m in M] for l in L]
r = np.array(r)

# Zmienna decyzyjna jeśli schron
x = [[1 for _ in range(h)] for _ in range(s + e)]


# Inicjalizacja QUBO - bierzemy pod uwagę już zmiennę typu slack
N = (s + e) * h + sum(slack_sizes)
Q = np.zeros((N, N))

def idx(i, n):
    return i * h + n
```

Funkcja celu: $$ \min \left( \sum_{i \in L} \sum_{n \in M} r_{in} \cdot x_{in} + \sum_{i \in L} \sum_{n \in M} c_i \cdot x_{in} \right) $$

```python
for i in range(s + e):
    for n in range(h):
        cost = c[i] + r[i][n]
        Q[idx(i, n), idx(i, n)] += cost
Q
```

Ograniczenia:
$$
\sum_{i \in L} x_{i,n} = 1 \quad \forall n \in M
$$

```python
for n in range(h):
    for i in range(s + e):
        q_i = idx(i, n)
        Q[q_i, q_i] += -P[0]
        for j in range(i + 1, s + e):
            q_j = idx(j, n)
            Q[q_i, q_j] += 2 * P[0]
Q
```

$$
\alpha \cdot \sum_{n \in M} x_{i,n} \cdot f_{n} \leq v_i \quad \forall i \in L
$$

```python
offset = (s + e) * h  # offset do slacków
for i in range(s + e):
    terms = []

    # Część x_{in} * α * f_n
    for n in range(h):
        terms.append((idx(i, n), alpha * f[n]))

    # Część slack: s_{ik} * 2^k
    for k in range(slack_sizes[i]):
        slack_idx = offset
        terms.append((slack_idx, 2 ** k))
        offset += 1

    # Stała prawa strona
    rhs = v[i]

    # Dodajemy do QUBO: (Ax + b)^2 = AᵀA + bᵀb + 2Aᵀb
    for u, coeff_u in terms:
        Q[u, u] += P[1] * coeff_u ** 2
        for v_, coeff_v in terms:
            if u < v_:
                Q[u, v_] += 2 * P[1] * coeff_u * coeff_v
        Q[u, u] += -2 * P[1] * coeff_u * rhs

    # Pomijamy: Q += P[1] * rhs^2 — bo stała nie wpływa na wynik
Q
```

$$
\sum_{i \in L} c_i \cdot x_i \leq p
$$

```python
# offset do slacków po poprzednich ograniczeniach
slack_budget_offset = (s + e) * h + sum(slack_sizes[:-1])  # sumujemy slacki #1–3
slack_bits = slack_sizes[-1]  # liczba bitów slacka budżetowego

terms = []

# Część x_{in} * c_i
for i in range(s + e):
    for n in range(h):
        q_idx = idx(i, n)
        terms.append((q_idx, c[i]))

# Część slack
for k in range(slack_bits):
    slack_idx = slack_budget_offset + k
    terms.append((slack_idx, 2 ** k))

# Prawa strona = budżet
rhs = p

# Dodaj do QUBO tak samo jak wcześniej
for u, coeff_u in terms:
    Q[u, u] += P[2] * coeff_u ** 2
    for v_, coeff_v in terms:
        if u < v_:
            Q[u, v_] += 2 * P[2] * coeff_u * coeff_v
    Q[u, u] += -2 * P[2] * coeff_u * rhs
Q
```

$$
x_{i,n} = 0 \quad \text{jeśli } r_{i,n} > d
$$
### Póki co zostawiam bez


```python
for i in range(s + e):
    for n in range(h):
        if r[i][n] > d:
            Q[idx(i, n), idx(i, n)] += P
Q
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from torch_geometric.data import Data
from torch.optim import Adam

class QUBOGCN(nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels, dropout=0.01):
        super(QUBOGCN, self).__init__()
        self.conv1 = GCNConv(in_channels, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, out_channels)
        self.dropout = dropout

    def forward(self, data):
        x, edge_index, edge_weight = data.x, data.edge_index, data.edge_weight
        x = self.conv1(x, edge_index, edge_weight=edge_weight)
        x = F.relu(x)
        x = F.dropout(x, p=self.dropout, training=self.training)
        x = self.conv2(x, edge_index, edge_weight=edge_weight)
        return torch.sigmoid(x).squeeze()
```

```python
def qubo_loss(probs, Q):
    probs = probs.view(-1, 1)  # (N, 1)
    Q = Q.detach().float().to(probs.device)
    return (probs.T @ Q @ probs).squeeze()
```

```python
def train_qubo_gnn(data, Q, model, optimizer, epochs=1000, tol=1e-4, patience=100, threshold=0.5):
    best_loss = float('inf')
    best_solution = None
    wait = 0

    if not isinstance(Q, torch.Tensor):
        Q = torch.tensor(Q, dtype=torch.float32)
    Q = Q.detach().to(model.parameters().__next__().device)

    for epoch in range(epochs):
        model.train()
        optimizer.zero_grad()
        probs = model(data)
        loss = qubo_loss(probs, Q)
        loss.backward()
        optimizer.step()

        if epoch % 100 == 0:
            print(f"Epoch {epoch}, Loss: {loss.item():.6f}")

        # Early stopping
        if loss.item() + tol < best_loss:
            best_loss = loss.item()
            best_solution = (probs >= threshold).float().detach()
            wait = 0
        else:
            wait += 1

        if wait >= patience:
            print(f"Early stopping at epoch {epoch}")
            break

    return best_solution, best_loss
```

```python
in_channels = 1
hidden_channels = 16
out_channels = 1
dropout = 0.1
learning_rate = 1e-3
epochs = 50000
tolerance = 1e-4
patience = 200
threshold = 0.5

model = QUBOGCN(in_channels, hidden_channels, out_channels, dropout)
model = model.to(torch.float32)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
data = data.to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

# ===== Trening =====
best_solution, best_loss = train_qubo_gnn(
    data=data,
    Q=Q,
    model=model,
    optimizer=optimizer,
    epochs=epochs,
    tol=tolerance,
    patience=patience,
    threshold=threshold
)

print("Najlepsze rozwiązanie (bitstring):")
print(best_solution.int().cpu().tolist())

print(f"Najlepsza wartość funkcji celu (QUBO): {best_loss:.4f}")
```
